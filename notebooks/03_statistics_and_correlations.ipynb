# Statistical Analysis

This notebook computes:
- Pearson correlations between entropy measures and behavioral variables
- Bonferroni-corrected p-values
- Calculates t-Test for Gender

Results are exported as CSV files.

```python
import numpy as np
import csv

from src.statistics import whole_brain, slice_entropies, t_test, whole_brain_gender, slice_entropies_gender
```

# Whole Brain Information

```python
whole_brain_cor = whole_brain()
print(whole_brain_cor)

for item in ['ages', 'fluid_Unadj', 'fluid_Adj', 'crystal_Unadj', 'crystal_Adj', 'total_Unadj', 'total_Adj']:
   for item2 in ['rest1_s200', 'rest2_s200', 'emotion_s200', 'gambling_s200', 'language_s200', 'motor_s200', 'relational_s200', 'social_s200', 'wm_s200']: 
        for item3 in ['std_raw', 'roc_raw_squared', 'se', 'fe', 'mse']:
            if item2 == 'emotion_s200' and item3 == 'mse':
                continue

            if whole_brain_cor[item][item2][item3][2][0] < 0.05: 
                print(item + " " + item2 + " " + item3 + " korrigiert:" + str(whole_brain_cor[item][item2][item3][2][0])) 
```

```python
csv_file = "WHOLE_BRAIN_COR_RESULTS.csv"

with open(csv_file, 'w', newline='') as file:
    fieldnames = ['Item', 'Durchgang', 'Correlation', 'Entropy']

    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()

    for item, item_results in whole_brain_cor.items():
        for durchgang, durchgang_results in item_results.items() :
            for entropy, entropie_results in durchgang_results.items():
                writer.writerow({'Durchgang': durchgang, 'Item': item, 'Entropy': entropy, 'Correlation': entropie_results})
```
